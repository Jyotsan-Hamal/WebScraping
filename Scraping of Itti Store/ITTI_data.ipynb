```python
gaming = []
Price = []
old_price = []
Laptop_Data = pd.DataFrame()
for j in range(1,16):
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    url = 'https://itti.com.np/gaming-laptops-nepal?p={}'.format(j)
    soup = BeautifulSoup(webpage,'lxml')
    webpage = requests.get(url,headers=headers).text
    Laptops = soup.find_all('li',class_='item product product-item')
    for i in Laptops:
        gaming.append(i.find_all('a',class_='product-item-link')[0].text.strip())
        Price.append(i.find_all('span',class_='price')[0].text.strip())
    d = {'Model' :gaming,'Price':Price}
    df = pd.DataFrame(d)
    Laptop_Data = Laptop_Data.append(df,ignore_index=True)
    
```

```python
Laptop_Data.to_csv('ITTI.csv')
```
