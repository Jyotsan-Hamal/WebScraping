```python
final = pd.DataFrame()
price = []
Name = []
Author = []

for j in range(1,35):
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    url = 'https://www.bookdepository.com/bestsellers?page={}'.format(j)
    webpage = requests.get(url,headers=headers).text
    soup = BeautifulSoup(webpage,'lxml')
    Book = soup.find_all('div',class_='book-item')
    for i in Book:
        Name.append(i.find_all('h3',class_='title')[0].text.strip())
        price.append(i.find_all('p',class_='price')[0].text.strip())
        Author.append(i.find_all('p',class_='author')[0].text.strip())
    d = {'Name':Name,'Author':Author,'Price':price}
    df = pd.DataFrame(d)
    final = final.append(df,ignore_index=True)
    
    
```

```python
final.to_csv('BookDataset.csv')
```
